```python
import requests
from bs4 import BeautifulSoup

url = 'https://wiki.python.org.br/AprendaMais'
requisicao = requests.get(url)
extracao = BeautifulSoup(requisicao.text, 'html.parser')
#Mostra a página completa
print(extracao.text.strip())
#OU
# print(extracao.text) - Ese é melhor

```

```python
#Encontra e mostra todos os títulos 'h2' e parágrafos 'p'
for linha_texto in extracao.find_all('h2'):
    titulo = linha_texto.name
    print(titulo)

# Filtrar exibição por TAG
for linha_texto in extracao.find_all('h2'):
    titulo = linha_texto.text.strip()
print('Título:', titulo)
    
for linha_texto in extracao.find_all('p'):
    titulo = linha_texto.text.strip()
print('Parágrafo:', titulo)    
```

```python
# # Desafio: 
# # filtrar tags ['h2', 'p']
# # Contar quantos h2 e p existem no documento (linha_texto.name == tag)
# # Contar a quantidade de títulos e parágrafos
contar_titulos = 0
contar_paragrafos = 0

for linha_texto in extracao.find_all(['p', 'h2']):
    if linha_texto.name == 'h2':
        contar_titulos += 1
        
    elif linha_texto.name == 'p':
        contar_paragrafos += 1
    
print('Total de Títulos: ', contar_titulos)
print('Total de Parágrafos: ', contar_paragrafos)
```

```python
# Desafio: 
# filtrar tags ['h2', 'p']
# Contar quantos h2 e p existem no documento (linha_texto.name == tag)

# Contar a quantidade de títulos e parágrafos
contar_titulos = 0
contar_paragrafos = 0

for linha_texto in extracao.find_all(['p', 'h2']):
    if linha_texto.name == 'h2':
        contar_titulos += 1
        
    elif linha_texto.name == 'p':
        contar_paragrafos += 1
    
print('Total de Títulos: ', contar_titulos)
print('Total de Parágrafos: ', contar_paragrafos)
```

```python
#Exibir somente o texto das tags 'h2' e 'p':

for linha_texto in extracao.find_all(['h2', 'p']):
    if linha_texto.name == 'h2':
        titulo = linha_texto.text.strip()
        print('Título: \n' , titulo)
    
    elif linha_texto.name == 'p':
        paragrafo = linha_texto.text.strip()
        print('Parágrafo: \n', paragrafo)
```

```python
#Exibir TAGS aninhadas:
for titulo in extracao.find_all('h2'):
    print('Título: \n', titulo.text.strip())
    for link in titulo.find_next_siblings('p'):
        for a in link.find_all('a', href=True):
            print('Texto Link: ', a.text.strip(), ' | URL: ', a['href'])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas

print('Request: ')
response = requests.get('https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/')
print(response.text[:600])
print("\n")

print('BeautifulSoup: ')
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify()[:1000])
print("\n")

print('Pandas: ')
url_dados = pandas.read_html('https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/')
print(url_dados[0].head(10))
```
